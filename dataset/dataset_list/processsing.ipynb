```python
# list all files in a folder and change its name to a new name using format 

import os
import re

def rename_files(folder, extract_name_regex, new_format):
    for filename in os.listdir(folder):
        print(filename)
        if os.path.isfile(os.path.join(folder, filename)):
            matches = re.findall(extract_name_regex, filename)
            if len(matches) == 0:
                print("No match found for file: ", filename)
                continue

            new_filename = new_format.format(matches[0])
            os.rename(os.path.join(folder, filename), os.path.join(folder, new_filename))
        

rename_files("ClipReady/04_LiTS/img", r"volume-(\d+).nii", "liver_{0}.nii")
```

```python
# list all subfolders, each will have two files one is image and the other is label i.e. imagingnii.gz and segmentation.nii.gz split them into two folders
# use last4 dgit of the folder name to rename the files, it should look like this:
# 05_KiTS/img/img0004.nii.gz	05_KiTS/label/label0004.nii.gz

def split_files(folder):
    for subfolder in os.listdir(folder):
        if os.path.isdir(os.path.join(folder, subfolder)):
            print(subfolder)
            for filename in os.listdir(os.path.join(folder, subfolder)):
                if os.path.isfile(os.path.join(folder, subfolder, filename)):
                    if "imaging" in filename:
                        os.rename(os.path.join(folder, subfolder, filename), os.path.join(folder, "img", f"img{subfolder[-4:]}.nii.gz"))
                    elif "segmentation" in filename:
                        os.rename(os.path.join(folder, subfolder, filename), os.path.join(folder, "label", f"label{subfolder[-4:]}.nii.gz"))
split_files("Kidney/kits19/data")
```

```python
# for list of folder
# add ".gz" to all files in the folder

def add_gz(folder):
    for filename in os.listdir(folder):
        if os.path.isfile(os.path.join(folder, filename)):
            os.rename(os.path.join(folder, filename), os.path.join(folder, filename + ".gz"))
add_gz("ClipReady/04_LiTS/img")
```

```python
# check for PAOT.txt integrity (i.e check if all neccessary files are exists, and delete the extra files)
# the PAOT.txt will have 2 columns, the first column is the image and second is the label
# check if the image exists, we can safely ignore the label, instead check if the post_label_exists 
# pao file should look like this:
# 13_AbdomenCT-12organ/img/Organ12_0001_0000.nii.gz	13_AbdomenCT-12organ/label/Organ12_0001.nii.gz

import os
import re
def check_pao_file(path,dataset_folder, skip_data_sets = [],check_label=False):
    with open(path) as f:
        for line in f:
            image, label = line.strip().split()
            sub_folder = image.split("/")[0]
            if sub_folder in skip_data_sets:
                continue
            if check_label and not os.path.exists(os.path.join(dataset_folder,label)):
                print("Label not found: ", label)
            if not os.path.exists(os.path.join(dataset_folder,image)):
                print("Image not found: ", image)
            # parts = image.split(os.sep)
            # if parts[1] == "img":
            #     parts[1] = "post_label"
            # else:
            #     parts[1] = "post_labelsTr"
            # parts[-1] = f"{parts[-1].split('.')[0]}.nii.gz"
            # label_path = os.sep.join(parts)
            post_label = label.replace("label", "post_label")
            post_label = post_label.replace(".nii.gz", ".h5")
            
            if not os.path.exists(os.path.join(dataset_folder,post_label)):
                print("Post Label not found: ", post_label)
check_pao_file("PAOT.txt", "/data/clip/ClipReady", ["02_TCIA_Pancreas-CT","03_CHAOS","12_CT-ORG","13_AbdomenCT-12organ"],True)
```

```python
p1 = "/data/clip/ClipReady/01_Multi-Atlas_Labeling/label/label0001c.nii.gz"
p2 = "/data/clip/ClipReady/01_Multi-Atlas_Labeling/label/label0001.nii.gz"

# load the two file above using nii.babel and compare if they have same size and data
import numpy as np
import nibabel as nib
def compare_nii(p1, p2):
    n1 = nib.load(p1)
    n2 = nib.load(p2)
    if n1.shape != n2.shape:
        print("Shape not equal")
    if not np.allclose(n1.get_fdata(), n2.get_fdata()):
        print("Data not equal")
    return n1,n2

n1,n2 = compare_nii(p1, p2)
```
