```python
# remote all line starts with certain element in a file

exclude = ['02', '03', '11','12','13']
def remove_line(file, exclude):
    with open(file, 'r') as f:
        lines = f.readlines()
    with open(file, 'w') as f:
        for line in lines:
            if not any(line.startswith(e) for e in exclude):
                f.write(line)
remove_line('PAOT.txt', exclude)

# create 3 files by removing lines starts with certain elements

def remove_line(data, exclude,new_data, data_folder = '/data/clip/ClipReady'):
    suffix = ['train', 'test', 'val']
    for s in suffix:
        with open(f'{data}_{s}.txt', 'r') as f:
            lines = f.readlines()
        with open(f'{new_data}_{s}.txt', 'w') as f:
            for line in lines:
                image, label = line.strip().split()
                if not os.path.exists(f'{data_folder}/{image}'):
                    print(f'{data_folder}/{image} does not exist remove line')
                    continue
                if not os.path.exists(f'{data_folder}/{label}'):
                    print(f'{data_folder}/{label} does not exist remove line')
                    continue
                if not any(line.startswith(e) for e in exclude):
                    f.write(line)
remove_line('PAOT', exclude, 'PAOT_TEMP')
```
